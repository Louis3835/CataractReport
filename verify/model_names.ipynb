```python
# Cell 1: Load YOLO models and extract class names
import os
from ultralytics import YOLO
import pandas as pd

# ================= 路径配置 =================
MODEL_CLS_PATH = "/home/zack/Desktop/workspace/zhaoubuntu/Segmentation/Yolo/runs/classify/train3/weights/best.pt"
MODEL_SEG_PATH = "/home/zack/Desktop/workspace/zhaoubuntu/Segmentation/Yolo/runs/segment/train9/weights/best.pt"
# ============================================

print("⏳ 正在载入 YOLO 专家模型权重...")
try:
    model_cls = YOLO(MODEL_CLS_PATH)
    model_seg = YOLO(MODEL_SEG_PATH)
    print("✅ 模型载入成功！\n")
except Exception as e:
    print(f"❌ 载入模型失败，请检查路径是否正确。\n错误信息: {e}")

# 提取分类模型的标签 (Surgery Phases)
cls_names = model_cls.names
df_cls = pd.DataFrame(list(cls_names.items()), columns=['Class ID', 'Phase Name (Surgery Phase)'])

# 提取分割模型的标签 (Tools & Tissues)
seg_names = model_seg.names
df_seg = pd.DataFrame(list(seg_names.items()), columns=['Class ID', 'Object Name (Tool/Tissue)'])

# ---- 漂亮的美化输出 ----
print("==================================================")
print(f" 1. 分类模型包含的标签总数 (Phases): {len(df_cls)}")
print("==================================================")
display(df_cls.style.set_properties(**{'text-align': 'left'}).hide(axis='index'))

print("\n==================================================")
print(f" 2. 分割模型包含的标签总数 (Tools/Tissues): {len(df_seg)}")
print("==================================================")
display(df_seg.style.set_properties(**{'text-align': 'left'}).hide(axis='index'))
```
